```python
#Realizar o destaque (highlight).
primeiras_20_linhas = df.head(20)

valores_maximos = primeiras_20_linhas.max()
valores_minimos = primeiras_20_linhas.min()

df_maximos = pd.DataFrame(valores_maximos).transpose()
df_minimos = pd.DataFrame(valores_minimos).transpose()

def highlight_max_min(s):
    is_max = s == valores_maximos
    is_min = s == valores_minimos
    styles = [f'background-color: {"blue" if v else "pink"}' for v in (is_max | is_min)]
    return styles

df_maximos_styled = df_maximos.style.apply(highlight_max_min, axis=1)
df_minimos_styled = df_minimos.style.apply(highlight_max_min, axis=1)

df_maximos_styled
df_minimos_styled
```

```python
#Criei um gráfico heatmap para a análise de correlação do item 16
correlation_matrix = df.corr(method='pearson')
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Matriz de Correlação (Método Pearson)')
plt.show()
```

```python
#Criei um scatterplot com interação para o par de variáveis com maior correlação.
fig = px.scatter(df, x='acidez volátil', y='acidez fixa', text=df.index)
fig.update_xaxes(title_text='acidez volátil')
fig.update_yaxes(title_text='acidez fixa')
fig.update_layout(title='Acidez Volátil x Acidez Fixa')
fig.update_traces(textposition='top center')

fig.update_layout(hovermode='closest')

fig.show()
```

```python
#Criei um correlograma.
correlation_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.set(style="white")
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(correlation_matrix, mask=mask, cmap=cmap, annot=True, fmt=".2f", square=True)
plt.title("Correlograma das Variáveis")
plt.show()
```

```python
#Realizei a análise univariada com um histograma para uma variável numérica (acido)
variavel_numerica = 'pH'
plt.figure(figsize=(8, 6))
plt.hist(df[variavel_numerica], bins=20, color='blue', alpha=0.7)
plt.title(f'Histograma da Variável {variavel_numerica}')
plt.xlabel(variavel_numerica)
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
#Apresentei em apenas um gráfico vários histogramas.
variaveis_numericas = ['álcool', 'pH', 'cloretos']
plt.figure(figsize=(12, 6))
for i, variavel in enumerate(variaveis_numericas, 1):
    plt.subplot(1, len(variaveis_numericas), i)
    sns.histplot(df[variavel], bins=20, kde=True, color='blue')
    plt.title(f'Histograma de {variavel}')
    plt.xlabel(variavel)
    plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

```python
#Boxplots
plt.figure(figsize=(29, 8))
num_colunas_subplot = 3
for i, col in enumerate(df.columns):
    if df[col].dtype != 'object':
        plt.subplot((len(df.columns) // num_colunas_subplot) + 1, num_colunas_subplot, i + 1)
        df[col].plot(kind='box', vert=False)
        plt.title(f'Boxplot de {col}')
        plt.xlabel(col)
        plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
#Gráfico de densidade para uma variável numérica.
variavel_numerica = 'açúcar residual'
plt.figure(figsize=(8, 6))  # Ajuste o tamanho da figura conforme necessário
df[variavel_numerica].plot(kind='density', title=f' {variavel_numerica}')
plt.xlabel(variavel_numerica)
plt.ylabel('Densidade')
plt.grid(True)
```

```python
#Gráfico de densidade para mais de uma variável numérica.

variaveis_numericas = ['álcool', 'densidade', 'pH', 'cloretos', 'acidez volátil']
plt.figure(figsize=(10, 6))

for variavel in variaveis_numericas:
    df[variavel].plot(kind='density', label=variavel)

plt.title('Gráfico de Densidade para Variáveis Numéricas')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)

plt.show()
```

```python
#Gráfico de barras ou colunas para exibir também o resultado do item 11. (sem outliers)
intervalos = pd.cut(df['pH'], bins=np.arange(2.5, 4.1, 0.1))
contagem_pH_agrupado = intervalos.value_counts().sort_index()
plt.figure(figsize=(10, 6))
contagem_pH_agrupado.plot(kind='bar', color='skyblue')
plt.title('Contagem de Valores Agrupados da Coluna "pH"')
plt.xlabel('Intervalos de pH')
plt.ylabel('Contagem')
plt.grid(axis='y')

plt.show()
```

```python
#Gráfico de pizza ou setores para uma variável categórica com porcentagens.
variavel = "qualidade"
contagem_valores = df[variavel].value_counts()
plt.figure(figsize=(8, 9))
plt.pie(contagem_valores, labels=contagem_valores.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Tipos de vinho "{variavel}"')
plt.axis('equal')
plt.show()
```
